```python
def retest_func(a, T):
    '''
    a is the json response from foo
    T is the original list of i'ds (t)
    '''
    # GET LIST OF ERRONOUS IDs
    error_list = []
    for i in a.json()['field']['Details']:
        if len(i['message'])==81:
            er = i['message'][34:-27]
        elif len(i['message'])==68 or 70:
            er = i['message'][34:-14]
        else: 
            print('the size of this error id does not fit predetermined solutions')
        error_list.append(er)
    print(len(error_list), error_list)
    # REMOVE ERRORS FOR INPUT
    for i in error_list:
        T.remove(i)

    # RUN DOWNLOAD AGAIN
    a = foo(T, collection_name, 1)
    return a
```
